
# ⚙️ Orquestração e Agendamento (Workflows)
## 🇧🇷 Português
### 🎯 Estratégia de Atualização
Para este projeto, configurei um **Databricks Workflow** que executa todo o pipeline uma vez por semana.
### 🛠️ Detalhes Técnicos:
* **Idempotência:** Como os dados são para portfólio, utilizei o comando `CREATE OR REPLACE`. Isso evita duplicidade de dados, pois a tabela é reconstruída a cada execução (Overwrite).
* **Dependências:** O Job respeita a ordem da arquitetura: Primeiro a Bronze, depois a Silver e, por fim, a Gold.

# ⚙️ Orchestration and Scheduling (Workflows)
## 🇺🇸 English 
### 🎯 Update Strategy
For this project, I configured a **Databricks Workflow** to run the entire pipeline once a week.

### 🛠️ Technical Details:
* **Idempotency:** Since this is a portfolio project, I used the `CREATE OR REPLACE` command. This prevents duplicate data because the table is rebuilt every time the job runs (Overwrite).
* **Dependencies:** The Job follows the architecture order: First Bronze, then Silver, and finally Gold.
