# 🥈 Step 2: Silver - Data Processing via SQL
## 🇺🇸 English
### 🎯 What does this script do?
This code performs the same transformations as the PySpark version, but uses **SQL**. It reads data from the **Bronze** layer and creates a refined table in the **Silver** layer.

### 🏗️ Business Rules applied:
* **Standardization:** Renaming columns to Portuguese.
* **Classification:** Using the `CASE WHEN` command to create `faixa_etaria` (age group) and `is_heavy_user` columns.
* **Conversion:** Transforming the academic impact column into numeric values (0 and 1).
* **Audit:** Including the date and time of the processing.


```python
%sql
-- Criando ou substituindo a tabela na camada Silver
CREATE OR REPLACE TABLE silver.social_media.social_media_impact AS
SELECT
    Student_ID AS id_estudante,
    Age AS idade,
    -- Criando Faixa Etária
    CASE 
        WHEN Age < 20 THEN '18-19'
        WHEN Age BETWEEN 20 AND 22 THEN '20-22'
        ELSE '23+' 
    END AS faixa_etaria,
    Gender AS genero,
    Country AS pais,
    CAST(Avg_Daily_Usage_Hours AS FLOAT) AS horas_uso_diario,
    -- Criando Indicador de Heavy User (Boolean)
    CASE 
        WHEN Avg_Daily_Usage_Hours > 6 THEN true -- ou simplesmente TRUE
        ELSE FALSE 
    END AS is_heavy_user,
    Most_Used_Platform AS plataforma_principal,
    -- Transformando Yes/No em Binário (0/1)
    CASE 
        WHEN Affects_Academic_Performance = 'Yes' THEN 1 
        ELSE 0 
    END AS impacto_academico_binario,
    Sleep_Hours_Per_Night AS horas_sono,
    Mental_Health_Score AS score_saude_mental,
    current_timestamp() AS data_processamento
FROM 
    bronze.social_media.social_media_impact;

-- Aplicando remoção de duplicatas (Opcional via SQL)
-- Nota: O SQL acima criará a tabela. Se houver duplicatas idênticas, 
-- o ideal é usar um DISTINCT após o SELECT.
```
