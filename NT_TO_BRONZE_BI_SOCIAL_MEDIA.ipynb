# 📥 Etapa 1: Ingestão de Dados (CSV para Bronze)
## 🇧🇷 Português
### 🎯 O que este código faz?
Este script é responsável por pegar os dados brutos que chegaram em um arquivo **CSV** e salvá-los de forma organizada em uma tabela **Delta** na nossa primeira camada da arquitetura (Bronze).

### 🏗️ Explicação do Passo a Passo:
1.  **Ler o arquivo:** O código acessa a pasta de entrada (`Volumes`) e abre o arquivo `dados.csv`.
2.  **Identificar colunas:** Ele lê os nomes das colunas (`header`) e tenta entender sozinho quais campos são números e quais são textos (`inferSchema`).
3.  **Salvar na Bronze:** O dado é gravado na tabela `bronze.sistema.tabela_bruta`.
    * **Delta:** Usamos esse formato porque ele é mais seguro e rápido.
    * **Append:** Esta opção adiciona novos dados ao final da tabela sem apagar o que já existe.

# 📥 Step 1: Data Ingestion (CSV to Bronze)
## 🇺🇸 English
### 🎯 What does this code do?
This script takes raw data from a **CSV** file and saves it into a **Delta** table in our first layer (Bronze).

### 🏗️ Step-by-Step Explanation:
1.  **Read the file:** The code goes to the input folder (`Volumes`) and opens the `dados.csv` file.
2.  **Identify columns:** It reads the column names (`header`) and automatically identifies which data are numbers or text (`inferSchema`).
3.  **Save to Bronze:** The data is saved into the table `bronze.sistema.tabela_bruta`.
    * **Delta:** We use this format because it is more secure and fast.
    * **Append:** This option adds new data to the end of the table without deleting old data.

```python
# Reading CSV
df_raw = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/bronze/social_media/raw_data/Social_media_impact_on_life.csv")

# Write delta table in Bronze Schema (Modificado para overwrite)
df_raw.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze.social_media.social_media_impact")
```
